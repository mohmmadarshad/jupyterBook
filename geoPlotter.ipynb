```python
x=fo.FeatureGroup(name='My Map')
```

```python
x.add_child(fo.Marker(location=[30.228175,78.803978],popup='this is my loaction',icon=fo.Icon(color='blue')))
```

```python
map.add_child(x)
```

```python
for lat,log,loc in ([14,17,'Dakar, Senegal'],[13,130,'Darwin, Australia'],[11,43,'Djibouti, Djibouti'],[53,6,'Dublin, Ireland']):
    x.add_child(
        fo.Marker(location=[lat,log],
        popup=loc,icon=fo.Icon(color='black'))
        )
```

```python
map.add_child(x)
```
