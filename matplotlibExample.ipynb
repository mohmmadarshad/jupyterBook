```python
# fig1=plt.figure()
# axes1=fig1.add_axes([1,1,2,2])
fig,axes=plt.subplots(2,2,figsize=(8,6))


axes[(1,0)].set_xlabel('x')
axes[(1,0)].set_ylabel('y')
axes[(1,0)].set_title('graph 1 x 0 ')
axes[(1,0)].plot(x,z,color='r',linewidth=12)
axes[(0,0)].set_xlabel('x')
axes[(0,0)].set_ylabel('y')
axes[(0,0)].set_title('graph 0 x 0 ')
axes[(0,0)].plot(z,x, color='b',alpha=0.8,marker='o',markersize=9,markerfacecolor='g'
)
axes[(1,1)].set_xlabel('x')
axes[(1,1)].set_ylabel('y')
axes[(1,1)].set_title('graph 1 x 1 ')
axes[(1,1)].plot(x,y,color='g',linestyle='--')
axes[(0,1)].set_xlabel('x')
axes[(0,1)].set_ylabel('y')
axes[(0,1)].set_title('graph 0 x 1 ')
axes[(0,1)].plot(y,x,linestyle=':')

plt.tight_layout()

```
